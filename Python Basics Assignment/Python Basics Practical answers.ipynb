```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def demonstrate_mutable_immutable(mutable_arg, immutable_arg):
    
    # Modifyingmutable argument (list)
    mutable_arg.append(100)

    # Modifying immutable argument (integer)
    immutable_arg += 10

    print("Inside function: ")
    print("Mutable argument: ", mutable_arg)
    print("immutable argument: ", immutable_arg)

# Calling the function 
my_list = [1,2,3]
my_number = 20

demonstrate_mutable_immutable(my_list, my_number)

print("\nOutside function: ")
print("Mutable argument: ", my_list)
print("immutable argument: ", my_number)
```

```python
# 14. write a program to reverse a string using a while loop

text = input("Enter a string: ")
i = len(text) - 1
reversed_text = ""

while i >= 0:
    reversed_text += text[i]
    i -= 1

print(reversed_text)
```

```python
# 15. Factorial using While loop

num = int(input("Enter a number: "))
factorial = 1

while num > 0:
    factorial *= num
    num -= 1

print("Factorial: ", factorial)
```
